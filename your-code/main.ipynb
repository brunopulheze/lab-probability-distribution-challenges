# Challenge 1 - Plot Probability Mass Function (PMF)

#### Background knowledge

[PMF](https://en.wikipedia.org/wiki/Probability_mass_function) shows the probability distribution of a **discrete random variable**. A [discrete random variable](https://en.wikipedia.org/wiki/Random_variable#Discrete_random_variable) has random numeric values that are not continuous. For example, the number of people in a household can only be integers but not floats. Therefore the number of people in a household is a discrete variable.

#### Question: We assume that the probability of clicking an Ad in a Youtube video is 0.15. We have a sample of 5 people who wathched the video and we want to plot the PMF for a binomial distribution.

*Hint: use binom from `scipy.stats.binom`. Your output should look like below:*

![binom 10](ch-2.png)

```python
# Parameters for the binomial distribution
n = 5  # number of trials (people)
p = 0.15  # probability of clicking an Ad

# Possible number of clicks (successes)
x = np.arange(0, n+1)

# Binomial PMF
pmf = stats.binom.pmf(x, n, p)

# Plotting
plt.plot(x, pmf, color='skyblue')
plt.xlabel('Number of people clicking Ad')
plt.ylabel('Probability')
plt.title('Binomial PMF (n=5, p=0.15)')
plt.xticks(x)
plt.show()
```

#### Explain what you observe from the plot above

The plot shows the probability distribution for the number of people (out of 5) who click an Ad, given each has a 15% chance.
Most likely, 0 or 1 person will click the Ad, as these have the highest probabilities.

#### Now plot PMP with 50, 500, and 5000 visitors.

```python
# Plot PMF for n = 50, 500, and 5000 visitors
visitor_counts = [50, 500, 5000]
plt.figure(figsize=(15, 4))

for i, n_visitors in enumerate(visitor_counts, 1):
    x_vals = np.arange(0, n_visitors + 1)
    pmf_vals = stats.binom.pmf(x_vals, n_visitors, p)
    plt.subplot(1, 3, i)
    plt.plot(x_vals, pmf_vals, color='skyblue')
    plt.title(f'Binomial PMF (n={n_visitors}, p={p})')
    plt.xlabel('Number of people clicking Ad')
    plt.ylabel('Probability')
    plt.xlim(0, n_visitors)
    plt.tight_layout()

plt.show()
```

#### What did you notice from the distribution plots? Comment your findings.

As the number of visitors increases, the distribution becomes more concentrated around the mean and appears more symmetric, resembling a normal distribution. The probability of extreme outcomes (very few or very many clicks) decreases.

#### Reaserch the Poisson distribution. Write about your own understanding of the Poisson distribution.

The Poisson distribution gives the probability of a number of events happening in a fixed interval, assuming events occur independently at a constant average rate (). It's used for counts like website visits per day. The mean and variance are both .

#### A website has an average of 300 visits per day. What is the probability of getting 320 visitors in a day?

*Hint: use `scipy.stats.poisson`*

```python
# Probability of getting 320 visitors when average is 300
lambda_ = 300
prob_320 = stats.poisson.pmf(320, lambda_)
print(f"Probability of getting 320 visitors: {prob_320}")
```

#### What is the probability of getting 60 visits?

```python
# Probability of getting 60 visitors when average is 300
prob_60 = stats.poisson.pmf(60, lambda_)
print(f"Probability of getting 60 visitors: {prob_60}")
```
